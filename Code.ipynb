```python
# Load the datasets
import pandas as pd

# Loading EV Market dataset
ev_market_data = pd.read_csv('/content/drive/MyDrive/Assignment/Chaganti-Reddy EVMarket-India.csv')

# Loading Normalized Electric Car Data dataset
electric_car_data_norm = pd.read_csv('/content/drive/MyDrive/Assignment/ElectricCarData_Norm.csv')

# Loading IEA EV data - EV sales Historical Cars dataset
iea_ev_data = pd.read_csv('/content/drive/MyDrive/Assignment/IEA-EV-dataEV_salesHistoricalCars.csv')

# Loading Charging Stations Sanctioned dataset
charging_stations_data = pd.read_csv('/content/drive/MyDrive/Assignment/RS_Session_256_AU_2673_2.ii_.csv')

# Loading Operational Charging Stations dataset
operational_charging_stations_data = pd.read_csv('/content/drive/MyDrive/Assignment/RS_Session_257_AU_2368_B_3.csv')

# Loading Allocations for Development Projects dataset
allocations_data = pd.read_csv('/content/drive/MyDrive/Assignment/RS_Session_259_AU_1769_2.csv')

# Display the first few rows of each dataset to understand its structure
ev_market_data.columns, electric_car_data_norm.columns, iea_ev_data.columns, charging_stations_data.columns, operational_charging_stations_data.columns, allocations_data.columns
```

```python
ev_market_data_selected = ev_market_data[['Brand', 'Model', 'AccelSec', 'TopSpeed_KmH', 'Range_Km', 'Efficiency_WhKm', 'FastCharge_KmH', 'PowerTrain', 'PlugType', 'BodyStyle', 'Segment', 'Seats', 'PriceEuro']]
```

```python
electric_car_data_norm_selected = electric_car_data_norm[['Brand', 'Model', 'Accel', 'TopSpeed', 'Range', 'Efficiency', 'FastCharge', 'PowerTrain', 'PlugType', 'BodyStyle', 'Segment', 'Seats', 'PriceEuro']]
```

```python
iea_ev_data_selected = iea_ev_data[['region', 'category', 'parameter', 'mode', 'powertrain', 'year', 'unit', 'value']]
```

```python
charging_stations_data_selected = charging_stations_data[['Sl. No', 'Category', 'Expressways/Highways', 'EV Charging Stations Sanctioned']]
```

```python
operational_charging_stations_data_selected = operational_charging_stations_data[['Sl. no. ', 'State/UT', 'City', 'Operational Charging stations Under FAME-II']]
```

```python
allocations_data_selected = allocations_data[['Sl. No.', 'State/UT', 'Circle / Branch', '2020-21- Allocation', '2021-22 - Allocation', '2022-23 - Allocation']]
```

```python
# Display basic information about each dataset
for dataset in [ev_market_data_selected, electric_car_data_norm_selected, iea_ev_data_selected, charging_stations_data_selected, operational_charging_stations_data_selected, allocations_data_selected]:
    print(dataset.info())
```

```python
# Display descriptive statistics for numerical columns
for dataset in [ev_market_data_selected, electric_car_data_norm_selected, iea_ev_data_selected, charging_stations_data_selected, operational_charging_stations_data_selected, allocations_data_selected]:
    print(dataset.describe())
```

```python
# Handle missing values
for dataset in [ev_market_data_selected, electric_car_data_norm_selected, iea_ev_data_selected, charging_stations_data_selected, operational_charging_stations_data_selected, allocations_data_selected]:
    dataset.dropna(inplace=True)  # Example: Remove rows with missing values
```

```python
# Convert data types if necessary
ev_market_data_selected['AccelSec'] = pd.to_numeric(ev_market_data_selected['AccelSec'], errors='coerce')  # Example: Convert 'AccelSec' to numeric
```

```python
# Example: Create a scatter plot for 'TopSpeed_KmH' vs 'Range_Km'
import matplotlib.pyplot as plt
plt.scatter(ev_market_data_selected['TopSpeed_KmH'], ev_market_data_selected['Range_Km'])
plt.xlabel('Top Speed (Km/H)')
plt.ylabel('Range (Km)')
plt.title('Scatter Plot: Top Speed vs Range')
plt.show()
```

```python
# Example: Apply K-Means clustering on relevant features for segmentation
from sklearn.cluster import KMeans
features_for_clustering = ev_market_data_selected[['TopSpeed_KmH', 'Range_Km', 'PriceEuro']]
kmeans = KMeans(n_clusters=3, random_state=42)
ev_market_data_selected['Segment'] = kmeans.fit_predict(features_for_clustering)
```

```python
# Merge datasets based on common columns (e.g., state/UT)
merged_data = pd.merge(iea_ev_data_selected, operational_charging_stations_data_selected, how='inner', left_on='region', right_on='State/UT')
```

```python
# Handle missing values
merged_data.dropna(inplace=True)
```

```python
# Group by location and calculate sum of relevant columns
location_data = merged_data.groupby('State/UT').agg({'value': 'sum', 'Operational Charging stations Under FAME-II': 'sum'})
```

```python
# Drop null values or impute them with meaningful values

# EV Market Data
ev_market_data_selected.dropna(inplace=True)

# Normalized Electric Car Data
electric_car_data_norm_selected.dropna(inplace=True)

# IEA EV Data - EV Sales Historical Cars
iea_ev_data_selected.dropna(inplace=True)

# Charging Stations Sanctioned
charging_stations_data_selected.dropna(inplace=True)

# Operational Charging Stations
operational_charging_stations_data_selected.dropna(inplace=True)

# Allocations for Development Projects
allocations_data_selected.dropna(inplace=True)
```
