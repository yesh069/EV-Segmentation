```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the dataset
df = pd.read_csv("/content/drive/MyDrive/Assignment/Book1.csv")

# Step 3: Explore the dataset
print(df.head())  # Display the first few rows of the dataframe
print(df.info())  # Get information about the dataframe (data types, missing values, etc.)

# Step 4: Data cleaning and preprocessing (if needed)
# Check for missing values and handle them if necessary
print(df.isnull().sum())  # Check for missing values in each column

# Step 5: Exploratory Data Analysis (EDA)
# Visualize distribution of numerical features
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in numeric_cols:
    plt.hist(df[col])
    plt.title(col)
    plt.show()

# Visualize relationships between features (e.g., scatter plots, correlation matrix)
plt.scatter(df['AccelSec'], df['Range_Km'])
plt.xlabel('Acceleration (sec)')
plt.ylabel('Range (Km)')
plt.title('Acceleration vs. Range')
plt.show()

# Step 6: Segmentation Analysis (if required)
# Segment the data based on certain criteria (e.g., demographic, psychographic)
# Analyze each segment separately and draw insights

# Example: Segmenting by BodyStyle
body_style_counts = df['BodyStyle'].value_counts()
print(body_style_counts)

# Visualize segment distribution
plt.bar(body_style_counts.index, body_style_counts.values)
plt.xlabel('Body Style')
plt.ylabel('Count')
plt.title('Distribution of Vehicles by Body Style')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Summary statistics
print(df2.describe())

# Value counts and bar plots for categorical features
categorical_cols = df2.select_dtypes(include=['object']).columns
for col in categorical_cols:
    plt.figure(figsize=(8, 6))
    sns.countplot(x=col, data=df2)
    plt.title(f'Count of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()

# Boxplot for numerical features by categorical features
numeric_cols = df2.select_dtypes(include=['float64', 'int64']).columns
for col in numeric_cols:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Segment', y=col, data=df2)
    plt.title(f'{col} Distribution by Segment')
    plt.xlabel('Segment')
    plt.ylabel(col)
    plt.xticks(rotation=45)
    plt.show()

# Heatmap for correlation between numerical features
plt.figure(figsize=(10, 8))
sns.heatmap(df2.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
